## Base population

```python
pop = pd.read_csv(lien+"/pop-2012.csv",delimiter=';', decimal=".", encoding='utf-8')
pop["COM"] = pop["COM"].map(str)
pop["DEP"] = pop["DEP"].map(str)
pop
```

```python
pop = pop[["DEP","COM","LIBCOM","P12_POP0019","P12_POP","P12_POP65P","P12_POPH","C12_POP15P_CS1","C12_POP15P","C12_POP15P_CS2",
     "C12_POP15P_CS3","C12_POP15P_CS4","C12_POP15P_CS5","C12_POP15P_CS6","C12_POP15P_CS7","P12_POP_IMM","P12_POP_FR"]]
pop
```

```python
pop = pop.groupby(["DEP","COM","LIBCOM"])[["P12_POP0019","P12_POP","P12_POP65P","P12_POPH","C12_POP15P_CS1","C12_POP15P","C12_POP15P_CS2","C12_POP15P_CS3",
                                     "C12_POP15P_CS4","C12_POP15P_CS5","C12_POP15P_CS6","C12_POP15P_CS7","P12_POP_IMM","P12_POP_FR"]].sum().reset_index()
```

```python
pop['part_0019'] = pop["P12_POP0019"]*100/pop["P12_POP"]
pop['part_65P'] = pop["P12_POP65P"]*100/pop["P12_POP"]
pop["part_H"] = pop["P12_POPH"]*100/pop["P12_POP"] 
pop["part_agri"] = pop["C12_POP15P_CS1"]*100/pop["C12_POP15P"]
pop["part_art_comm"] = pop["C12_POP15P_CS2"]*100/pop["C12_POP15P"]
pop["part_cadre"] = pop["C12_POP15P_CS3"]*100/pop["C12_POP15P"]
pop["part_prof_inter"] = pop["C12_POP15P_CS4"]*100/pop["C12_POP15P"]
pop["part_employe"] = pop["C12_POP15P_CS5"]*100/pop["C12_POP15P"]
pop["part_ouvrier"] = pop["C12_POP15P_CS6"]*100/pop["C12_POP15P"]
pop["part_retraite"] = pop["C12_POP15P_CS7"]*100/pop["C12_POP15P"]
pop["rap_immig_francais"] = pop["P12_POP_IMM"]/pop["P12_POP_FR"]
```

```python
pop = pop[["DEP","COM","LIBCOM","P12_POP",'part_0019','part_65P',"part_H","part_agri","part_art_comm","part_cadre","part_prof_inter","part_employe","part_ouvrier","part_retraite","rap_immig_francais"]]
pop
```

## Base formation

```python
formation = pd.read_csv(lien+"/formation-2012.csv",delimiter=';', decimal=".", encoding='utf-8')
formation["COM"] = formation["COM"].map(str)
formation["DEP"] = formation["DEP"].map(str)
```

```python
formation["P12_POP15P"] = formation["P12_POP1517"]+formation["P12_POP1824"]+formation["P12_POP2529"]+formation["P12_POP30P"]
```

```python
formation = formation[["DEP","COM","LIBCOM","P12_POP15P","P12_NSCOL15P_DIPL0","P12_NSCOL15P_CEP","P12_NSCOL15P_BEPC","P12_NSCOL15P_CAPBEP","P12_NSCOL15P_BAC","P12_NSCOL15P_BACP2","P12_NSCOL15P_SUP"]]
formation
```

```python
formation = formation.groupby(["DEP","COM","LIBCOM"])[["P12_POP15P","P12_NSCOL15P_DIPL0","P12_NSCOL15P_CEP","P12_NSCOL15P_BEPC","P12_NSCOL15P_CAPBEP","P12_NSCOL15P_BAC","P12_NSCOL15P_BACP2","P12_NSCOL15P_SUP"]].sum().reset_index()
formation
```

```python
formation[[dep in ['0'+str(i) if i in [j for j in range(1,11)] else str(i) for i in range(1,20)] for dep in formation["DEP"]]]
```

```python
formation["part_nonscol_sdip"] = formation["P12_NSCOL15P_DIPL0"]*100/formation["P12_POP15P"]
formation["part_nonscol_cep"] = formation["P12_NSCOL15P_CEP"]*100/formation["P12_POP15P"]
formation["part_nonscol_bepc"] = formation["P12_NSCOL15P_BEPC"]*100/formation["P12_POP15P"]
formation["part_nonscol_capbep"] = formation["P12_NSCOL15P_CAPBEP"]*100/formation["P12_POP15P"]
formation["part_nonscol_bac"] = formation["P12_NSCOL15P_BAC"]*100/formation["P12_POP15P"]
formation["part_nonscol_bacp2"] = formation["P12_NSCOL15P_BACP2"]*100/formation["P12_POP15P"]
formation["part_nonscol_sup"] = formation["P12_NSCOL15P_SUP"]*100/formation["P12_POP15P"]
```

```python
formation = formation[["DEP","COM","LIBCOM","P12_POP15P",'part_nonscol_sdip','part_nonscol_cep',"part_nonscol_bepc","part_nonscol_capbep","part_nonscol_bac","part_nonscol_bacp2","part_nonscol_sup"]]
formation
```

```python
formation[[dep in ['0'+str(i) if i in [j for j in range(1,11)] else str(i) for i in range(1,20)] for dep in formation["DEP"]]]
```

```python
formation["COM"] = [str(com)[1:] if str(com)[0] == '0' else str(com) for com in formation["COM"]]
```

## Data merging

```python
from functools import reduce
data_frames_to_merge = [pop, 
                        formation[["COM","P12_POP15P",'part_nonscol_sdip','part_nonscol_cep',"part_nonscol_bepc","part_nonscol_capbep","part_nonscol_bac","part_nonscol_bacp2","part_nonscol_sup"]], 
                        activite[["COM","chomage1564","chomage1524","chomage2554","chomage5564","chomage1564H","chomage1564F","inactifs","Eleves_stag_nr","retraite_preretraite","actifs_occ_agri_exploi","actifs_occ_artis_comm","actifs_occ_cadres","actifs_occ_prof_interm","actifs_occ_employe","actifs_occ_ouvrier","salarie15P","salarieTP15P","FPouCDI","CDD","job_commune_resid","job_autredep_meme-reg","job_autrereg"]]]
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['COM'],
                                            how='left'), data_frames_to_merge)
df_merged
```

```python
mask_duplicated = df_merged[['DEP','COM','LIBCOM']].duplicated(keep=False)
df_merged[mask_duplicated]
```
