```python
id='1vEzkclLjoPjgI-NUjke5VvmOJc7-qAMi'
linke="https://drive.google.com/uc?export=download&id="+id

linke="/content/drive/MyDrive/Abstention et vacances scolaires/Base créées/final_iris.csv"
iris = pd.read_csv(linke,delimiter=',', decimal=".", encoding='UTF-8')
iris["key"]=iris["DEP"].map(str)+'.'+iris["COM"].map(str)
iris["key"]
```

```python
iris['key'] = [str(dep) +'.' + str(int(str(com)[len(str(dep)):])) for com,dep in zip(iris['COM'],iris['DEP'])]
iris[['COM','key']]

iris.loc[iris.DEP==69].key
```

```python
abstention.columns
```

```python
abstention_data = pd.merge(abstention,
         iris,
         on=['key'],how='inner')


abstention_data
```

```python
abstention_data = abstention_data.drop(columns=["Code du département","Code de la commune","Nom de la commune"])
abstention_data.rename(columns={"Libellé du département":"LIBDEP"},inplace=True)
abstention_data.columns
```

```python
abstention_data.to_csv("/content/drive/MyDrive/Abstention et vacances scolaires/Base créées/abstention_data.csv",encoding='UTF-8',index = False)
abstention_data
```
