```python
def clean(data):
  output = str(data).replace('\t','')
  output = str(output).replace(' ','')
  return int(output)
def clean2(data):
  output = str(data).replace('\t','')
  output = str(output).replace(',','.')
  return float(output)
```

```python
abstention['Inscrits2012'] = abstention['Inscrits2012'].map(clean)
abstention['Inscrits2007'] = abstention['Inscrits2007'].map(clean)
abstention['Inscrits2017'] = abstention['Inscrits2017'].map(clean)

abstention['abs2017'] = abstention['abs2017'].map(clean2)
abstention['abs2012'] = abstention['abs2012'].map(clean2)
abstention['abs2007'] = abstention['abs2007'].map(clean2)
```

```python
match_right_keys = []
k = 201
for i in range(551,568):
  if i !=560:
    match_right_keys.append(('13.'+str(i),'13.'+str(k)))
    k = k+1

match_right_keys.append(('69.123','69.381'))

k = 382
for i in range(1231,1239):
  match_right_keys.append(('69.'+str(i),'69.'+str(k)))
  k = k+1

match_right_keys
```

```python
for key in match_right_keys:
  abstention["key"].replace({key[0]:key[1]}, inplace=True)
```
